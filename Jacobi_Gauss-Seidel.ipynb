Jacobi method

```python
import math
import matplotlib.pyplot as plt
import numpy as np

a11 = 5.4
a12 = -6.2
a13 = -0.5
a21 = 3.4
a22 = 2.3
a23 = 0.8
a31 = 2.4
a32 = -1.1
a33 = 3.8
b1 = 6.5
b2 = -0.8
b3 = 1.8

def f2(x, z):
    y1 = (-a11*x - a13*z + b1)/a12
    return y1
def f1(y, z):
    x1 = (-a22*y - a23*z + b2)/a21
    return x1
def f3(x, y):
    z1 = (-a31*x - a32*y + b3)/a33
    return z1

if abs(a21)>abs(a22+a23) and abs(a33)>abs(a31+a32) and abs(a12)>abs(a11+a13):
    print("The iterative process is convergent ")
else:
    print("The iterative process is running out of steam")

x0 = 0
y0 = 0
z0 = 0
i = 1

e = 0.0001


log_list = []
it_list = []


condition = True

while condition:
    x1 = f1(y0,z0)
    y1 = f2(x0,z0)
    z1 = f3(x0,y0)
    
    e1 = abs(x0-x1);
    e2 = abs(y0-y1);
    e3 = abs(z0-z1);


    if (e1 > e2) and (e1 > e3):
        max = e1
        logarithm = math.log(max)
        log_list.append(logarithm)
    elif (e2 > e1) and (e2 > e3):
        max = e2
        logarithm = math.log(max)
        log_list.append(logarithm)
    else:
        max = e3
        logarithm = math.log(e1)
        log_list.append(logarithm)

    it_list.append(i)

    print(i, "iteration х =", x1, "value у =", y1, "value z =", z1, "\n", "delta х1 =",e1, "delta у1 =",e2, "delta z1 =",e3)
    i = i + 1
    x0 = x1
    y0 = y1
    z0 = z1
    condition = e1>e and e2>e and e3>e

    if i >= 100:
        break

plt.plot(it_list, log_list, 'b', marker = 'x') 
plt.show()


```
