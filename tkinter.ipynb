```python
import tkinter as tk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
```

```python
from tkinter import *
root = Tk()
root.wm_title("Graph of free fall")

class Fall(object):
    
    
    g = 9.8066 #the value of the gravitational constant
    
    def __init__(self, h, v0):
        self.__h = h
        self.__v0 = v0
        
    def y(self, t):
        return self.__h - (Fall.g * (t**2))/2
    
    def x(self, t):
        return self.__v0 * t
    
    def time_run(self):
        return np.sqrt(2 * self.__h / Fall.g)   
    
h = int(input("Enter the height: "))
v0 = int(input("Enter the velocity: "))

fall = Fall(h, v0)
t_run = fall.time_run()

figure_1 = plt.figure(figsize=(14, 7))
line2 = FigureCanvasTkAgg(figure_1, root)
line2.get_tk_widget().pack(side=tk.LEFT, fill=tk.BOTH)


t = np.linspace(0, t_run, 1000)
x = fall.x(t)
y = fall.y(t)

plt.xlabel('Axis x')
plt.ylabel('Axis y')

plt.plot(x, y, color='blue', linewidth=2)
plt.title("Graph of motion of a body thrown horizontally")


  
root.mainloop()
```
